```python
relevant_features=['Transaction_Amount','Average_Transaction_Amount','Frequency_of_Transactions']
target=['Is_Anomaly']

x=data[relevant_features]
y=data[target]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model=IsolationForest(contamination=0.2,random_state=42)
model.fit(x_train)
```

1. Feature and Target Selection:
   - relevant_features contains the columns used as input features (x).
   - target contains the column used as the target variable (y).

2. Splitting Data:
   - train_test_split splits the data into training and testing sets with 80% of the data used for training and 20% for testing.

3. Model Initialization:
   - IsolationForest is initialized to detect anomalies, with a contamination factor of 20%, meaning it expects 20% of the data to be anomalies.

4. Model Training:
   - model.fit(x_train) trains the Isolation Forest model using the training data.

This setup prepares the model to detect anomalies in transaction data based on the specified features.

```python
# Evaluate the model's performance
report = classification_report(y_test, y_pred_binary, target_names=['Normal', 'Anomaly'])
print(report)
```

```python
# Relevant features used during training
relevant_features = ['Transaction_Amount', 'Average_Transaction_Amount', 'Frequency_of_Transactions']

# Get user inputs for features
user_inputs = []
for feature in relevant_features:
    user_input = float(input(f"Enter the value for '{feature}': "))
    user_inputs.append(user_input)

# Create a DataFrame from user inputs
user_df = pd.DataFrame([user_inputs], columns=relevant_features)

# Predict anomalies using the model
user_anomaly_pred = model.predict(user_df)

# Convert the prediction to binary value (0: normal, 1: anomaly)
user_anomaly_pred_binary = 1 if user_anomaly_pred == -1 else 0

if user_anomaly_pred_binary == 1:
    print("Anomaly detected: This transaction is flagged as an anomaly.")
else:
    print("No anomaly detected: This transaction is normal.")
```
