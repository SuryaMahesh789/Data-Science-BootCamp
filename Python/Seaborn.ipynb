```python
#df = df.replace([np.inf, -np.inf], np.nan)
df = {'order_id': [1, 2, np.inf, 4, 5]}
df = pd.DataFrame(df)

# Replace infinite values with NaN
df = df.replace([np.inf, -np.inf], np.nan)

# Create the histogram
sns.histplot(df['order_id'])

```

```python
df['order_id'].unique()  # Unique values in a column
df['order_id'].value_counts()  # Count occurrences of each value
```

```python
sns.boxplot(x='order_id', data=df)#BOXPLOT
plt.show()
```

```python
sns.histplot(x='order_id',data=df)#HISTPLOT
plt.show()
```

```python
sns.scatterplot(x1='order_id',x2='order_date',data=df)#SCATTER PLOTS
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True)#CORRELATION MATRIX
plt.show()
```

```python
sns.pairplot(df)#PAIRPLOTS
plt.show()
```
